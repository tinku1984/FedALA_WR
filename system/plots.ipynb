```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

# Define file paths
file_paths = [
    '/home/tinku/nclab/PFLlib/results/Cifar10_pat_FedALA++_test_0_b_10.h5',
    '/home/tinku/nclab/PFLlib/results/Cifar10_pat_FedALA_test_0.h5',
    '/home/tinku/nclab/PFLlib/results/Cifar10_pat_FedALA++_test_0_b_20.h5',
    '/home/tinku/nclab/PFLlib/results/Cifar10_pat_FedALA++_test_0_b_50.h5'
    
    
]

# Labels for the legend
labels = [
    'FedALA',
    'FedALA++ (Buffer Size = 10)',
    'FedALA++ (Buffer Size = 20)',
    'FedALA++ (Buffer Size = 50)'
]

# Initialize the plot
plt.figure(figsize=(12, 8))

# Loop through each file, load the data, and plot
for file_path, label in zip(file_paths, labels):
    with h5py.File(file_path, 'r') as hf:
        rs_test_acc = np.array(hf.get('rs_test_acc'))  # Access test accuracy
    
    # Select the range between 1 to 100 rounds
    start_round = 1
    end_round = 100
    selected_rounds = range(start_round, end_round + 1)
    selected_acc = rs_test_acc[start_round:end_round + 1]
    
    # Plot the accuracy
    plt.plot(selected_rounds, selected_acc, label=label)

# Add labels, title, and legend
plt.title('Comparison of Federated Learning Test Accuracy', fontsize=16)
plt.xlabel('Communication Rounds', fontsize=14)
plt.ylabel('Test Accuracy', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import h5py
import matplotlib.pyplot as plt



# File paths for FedAvg and Incremental FL with EWC results
fedavg_path = '/home/tinku/nclab/results/200_Al_P1_Cifar10_pat_FedALA++_test_0.h5'
# fedavg_path = '/home/tinku/nclab/results/200_Cifar10_pat_FedALA_test_0.h5'
fedewc_path = '/home/tinku/nclab/results/200_Cifar10_pat_FedALA++_test_0.h5'

# Load FedAvg results
with h5py.File(fedavg_path, 'r') as hf:
    fedavg_acc = np.array(hf.get('rs_test_acc'))

# Load Incremental FL with EWC results
with h5py.File(fedewc_path, 'r') as hf:
    fedewc_acc = np.array(hf.get('rs_test_acc'))

# Select the range between 100 to 200 rounds
start_round = 0
end_round = 200
selected_rounds = range(start_round, end_round + 1)

# Slice the data for the selected range
fedavg_acc_selected = fedavg_acc[start_round:end_round + 1]
fedewc_acc_selected = fedewc_acc[start_round:end_round + 1]

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(selected_rounds, fedavg_acc_selected, label='FedALA', color='blue', linestyle='-')
plt.plot(selected_rounds, fedewc_acc_selected, label='FedALA++', color='red', linestyle='-')

# Add labels, title, legend, and grid
plt.title('Comparison of Test Accuracy (400 Rounds )', fontsize=16)
plt.xlabel('Communication Rounds', fontsize=14)
plt.ylabel('Test Accuracy', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
